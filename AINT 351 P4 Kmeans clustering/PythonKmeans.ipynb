```python
import numpy as np
import matplotlib.pyplot as plt
from functions import GenerateGaussianData
from functions import Kmeansclassifier
sample1= 10000
sample2= 10000
mean1 = np.matrix('-4; -1')
mean2 = np.matrix('3; 4')
sigma1 = np.matrix('0.75') 
sigma2 = np.matrix('2.0')
```

```python
testing1, target1 = GenerateGaussianData(sample1, mean1, sigma1, mean2, sigma2)
testing2, target2 = GenerateGaussianData(sample2, mean1, sigma1, mean2, sigma2)
```

```python
fidx = np.where(target1[0,:] == 1)
c1data = testing1[:,fidx[1]]
fidy = np.where(target2[0,:] == 0)
c2data = testing2[:,fidy[1]]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter([c1data[0,:]], [c1data[1,:]], c='b', marker="s")
ax1.scatter([c2data[0,:]], [c2data[1,:]], c='r', marker="o")
```

```python
declustered = np.hstack((c1data,c2data))
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter([declustered[0,:]], [declustered[1,:]], c='b', marker="s")
```

```python
[class1,class2]=Kmeansclassifier(declustered,2 )
```

```python
fig2 = plt.figure()
ax2=fig2.add_subplot(111)

ax2.scatter([class1[0,:]], [class1[1,:]], c='b', marker="s")
ax2.scatter([class2[0,:]], [class2[1,:]], c='r', marker="o")
```
